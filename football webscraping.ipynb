```python
import requests
```

```python
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
#request.get take the HTML of the page
data = requests.get(standings_url)
```

```python
#.text turn HTML into text (this is the code that render into web page)
data.text
```

```python
from bs4 import BeautifulSoup
```

```python
#initialize object
soup = BeautifulSoup(data.text)
```

```python
#select the data which is the whole table
#select(name of the table. class name), [0] means we only select the first one
standings_table = soup.select('table.stats_table')[0]
```

```python
#identify a tag inside our table, a element is the link in the table (not value but the link to the next page)
links = standings_table.find_all('a')
```

```python
#get href property from a element
links = [l.get("href") for l in links]
```

```python
#filter out links with /squads in front only
links = [l for l in links if '/squads/' in l]
```

```python
# add https://fbref.com{l} in front of the l in links
team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
team_urls
```

```python
#extract data from each l (each page from each a tag url)
team_url = team_urls[0]
data = requests.get(team_url)
```

```python
#list of year we want to scrap
#-1 is the order from the 2023 first
years = list(range(2023, 2020, -1))
```

```python
years
```

```python
all_matches = []
```

```python
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
import time

for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text)
    try:
        standings_table = soup.select('table.stats_table')[0]
        #find team link
        links = [l.get("href") for l in standings_table.find_all('a')]
        #filter out only wanted filter
        links = [l for l in links if '/squads/' in l]
        #add prefix
        team_urls = [f"https://fbref.com{l}" for l in links]
    except ValueError:
        continue
        
    #url of the previous season (button)
    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"
    
    for team_url in team_urls:
        #split the url in the only the name of the team (-1 means the name is at the back)
        #replace ("-Stats") with ""
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", "")
        
        team_data = requests.get(team_url)
        matches = pd.read_html(team_data.text, match="Scores & Fixtures")[0]
        
        soup = BeautifulSoup(data.text)
        links = [l.get("href") for l in standings_table.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        
        try:
            data = requests.get(f"https://fbref.com{links[0]}")
            shooting = pd.read_html(data.text, match="Shooting")[0]
            shooting.columns = shooting.columns.droplevel()
       
        except ValueError:
            continue
        
        #some teams shooting stat is not available, it is to say that just ignore it
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError:
            continue
            
        #filter only premier leage
        team_data = team_data[team_data["Comp"] == "Premier League"]
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        #make sure we dont get block from website
        time.sleep(2)
        
        
    
```
